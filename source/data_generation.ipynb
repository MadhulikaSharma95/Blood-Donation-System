```python
def phn():
    n = '0000000000'
    while '9' in n[3:6] or n[3:6]=='000' or n[6]==n[7]==n[8]==n[9]:
        n = str(random.randint(10**9, 10**10-1))
    return n[:3] + n[3:6] + n[6:]

profile = fake.profile()
name = profile['name']
age = random.randint(10, 90)
blood_group = profile['blood_group']
a = real_random_address()
address = a['address1']+a['address2']+', '+a['city']+', '+a['state']+', '+a['postalCode']
phone = int(phn())
# profile.replace("\n", ",")
profile
```

```python
donor_list = []
for i in range(3000):
    donor = defaultdict()
    profile = fake.profile()
    donor_id = 'd'+f'{i:04}'
    name = profile['name']
    age = random.randint(21, 65)
    blood_group = profile['blood_group']
    address = profile['address'].replace("\n", ", ")
    phone = int(phn())
    donor["donor_id"] = donor_id
    donor["name"] = name
    donor["age"] = age
    donor["Blood_group"] = blood_group
    donor["Address"] = address
    donor["contact_number"] = phone
    donor_list.append(donor)
```

```python
donor_db = pd.DataFrame(donor_list)
donor_db
```

```python
inventory_list = []
bg = ['A+', 'A-', 'B+', 'B-', 'AB+', 'AB-', 'O+', 'O-']
for i in range(0,1000):
    for b in bg:
        inventory = defaultdict()
        inventory['blood_bank_id'] = 'b'+f'{i:04}'
        inventory['blood_group'] = b
        inventory['units_available'] = random.randint(0, 100)
        inventory_list.append(inventory)
inventory_db = pd.DataFrame(inventory_list)
inventory_db
```

```python
blood_bank_list = []
i = 0
while i <1000:
    bb = defaultdict()
    r = real_random_address()
    blood_bank_id = 'b'+f'{i:04}'
    if 'city' in r.keys() and 'postalCode' in r.keys() and 'address1' in r.keys() and 'address2' in r.keys() and 'state'  in r.keys():
        name = r['city']+' Blood Center '+f'{i:03}'
        address = r['address1']+r['address2']+', '+r['city']+', '+r['state']+', '+r['postalCode']
        phone = int(phn())
        i += 1
        bb['blood_bank_id'] = blood_bank_id
        bb['name'] = name
        bb['address'] = address
        bb['phone'] = phone
        blood_bank_list.append(bb)
blood_bank_db = pd.DataFrame(blood_bank_list)
blood_bank_db
```

```python
recipient_data = []
for i in range(3000):
    # r_id = "rc"+f"{i:04}"
    name = " ".join(gen_data.create_name())
    age = random.randint(0, 99)
    blood_group = random.choice(bg)
    prob = np.random.random()
    if i%2 == 0:
        loc_hosp = random.choice(hosp_id)
        loc_ngo = np.nan
    else:
        loc_hosp = np.nan
        loc_ngo = random.choice(ngo_id)
        
    phnum = gen_data.create_phone()
    recipient_data.append((name, age, blood_group, loc_ngo, loc_hosp, phnum))
    
```

```python
request_data = []
for i in range(3000):
    # rq_id = "rq"+ f"{i:04}"
    blood_group = random.choice(bg)
    rc_ids = recipient_db[recipient_db["BloodGroup"] == blood_group]["ID"].values.tolist()
    patient_id = random.choice(rc_ids)
    d_ids = donor_db[donor_db["Blood_group"] == blood_group]["donor_id"].values.tolist()
    loc_obtained = random.choice(loc_ids)
    probability = np.random.random()
    
    if probability <= 0.7:
        donor_id_f = random.choice(d_ids)
        completed = True
        waitlist = 0
    else:
        donor_id_f = np.nan
        completed = False
        waitlist = random.randint(1, 10)
        
    request_data.append((patient_id, donor_id_f, blood_group, loc_obtained, waitlist, completed))
```

```python
random.shuffle(request_data)
```

```python
request_db = pd.DataFrame(request_data, columns=["PatientID", "DonorID", "BloodGroup", "Loc_Obtained", "Waitlist", "Completed"])
```
